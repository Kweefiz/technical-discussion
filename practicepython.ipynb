#5: List Overlap
---

```python
import random
def list_generate(count, lower_bound = 0, upper_bound = 100):
    return [random.randint(lower_bound, upper_bound) for i in range(count)]

a = list_generate(int(input("How many numbers for list a? ")))
b = list_generate(int(input("How many numbers for list b? ")))
c = [i for i in a if i in b]
print(a)
print(b)
print(c)
```

#6: String Lists

```python
inp = (input("Pick a word ")).lower()
word_check = [i for i in inp]
word_copy = word_check[::-1]
for i in word_check:
  if i != word_copy[word_check.index(i)]:
    print("Not a palindrome")
    break
else:
  print("It's a palindrome")
```

#9: Guessing Game One
---

```python
import random

randnum = random.randint(1,10)
count = 0
while True:
  user_number = int(input("Guess a number between 1 and 10 "))
  count += 1
  if user_number == randnum:
    print("You Win!")
    break
  elif user_number < randnum:
    print("Your guess was lower, try again")
  else:
    print("Your guess was higher, try again")
print(f"you took {count} tries!")
```

#13: Fibonacci
---

```python

def fibonacci():
  count = int(input("Choose how many numbers you want in the sequence "))
  i = 1
  if count == 0:
    fib = []
  elif count == 1:
    fib = [1]
  elif count == 2:
    fib = [1,1]
  elif count > 2:
    fib = [1,1]
    while i < (count-1):
      fib.append(fib[i]+fib[i-1])
      i += 1
  return fib
print(fibonacci())


```

#14: List Remove Duplicates
---

```python
#Loop Method
import random
count = int(input("Pick how many numbers to be in this list "))
a = [random.randint(1,10) for i in range(count)]
print(a)
no_Dupe = []
for i in a:
  if i not in no_Dupe:
    no_Dupe.append(i)
no_Dupe.sort()
print(no_Dupe)
```

```python
#Set Method
count = int(input("Pick how many numbers to be in this list "))
a = [random.randint(1,10) for i in range(count)]
print(a)
no_Dupe = set(a)
print(no_Dupe)
```

#15: Reverse Word Order
---

```python
#straightforward
name = input(" 'pick a sentence' ")
def reverseword(sentence):
  strings = sentence.split()
  reversed = strings[::-1]
  return  ' '.join(reversed)
reverseword(name)
```

```python
#comprehensive method
name = input(" Pick a sentence ")
def reverseword(sentence):
  return ' '.join(sentence.split()[::-1])
print(reverseword(name))
```

#16: Password Generator
---

```python
import random

while True:
  s = "abcdefghijklmnopqrstuvwxyz01234567890ABCDEFGHIJKLMNOPQRSTUVWXYZ!@#$%^&*()?"
  passlen = 8
  p = ''.join(random.sample(s,passlen))

  def check_upper(pw):
    global uppercase
    uppercase = False
    for i in pw:
      if i.isupper():
        uppercase = True
        break


  def check_lower(pw):
    global lowercase
    lowercase = False
    for i in pw:
      if i.islower():
        lowercase = True
        break

  def check_special(pw):
    global special
    special = False
    for i in pw:
      if i in "!@#$%^&*-?\/":
        special = True
        break

  def check_character(pw):
    global character
    character = False
    if len(pw)>=8:
      character = True

  check_upper(p)
  check_lower(p)
  check_special(p)
  check_character(p)

  print(p)




print(uppercase)
print(lowercase)
print(special)
print(character)
```
